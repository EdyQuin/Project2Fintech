# Analyzing text using Natural Language Processing techniques

## Analyze 'disaster' datasets in the form of tweets 

```python
# Necessary imports for WordCloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt #to display our wordcloud
import numpy as np #to get the color of our image
from PIL import Image #to load our image
# Pulling the tweets data set
text = open('./data/natdisastone.csv', mode="r", encoding="utf-8").read() # The data
stopwords = set(STOPWORDS)
```

## Analyze "tweets" using the 'wordcloud' technique

```python
# Show the catergorial data (tweets) in the string object to find the subject disaster
text
```

```python
# Use WordCloud to create a wordcloud of the tweets
wc = WordCloud(background_color="black",
               stopwords=stopwords,
               width=1600,
               height=1000,
               repeat=False,
               min_word_length=4,
               include_numbers=True
)
wc.generate(text)
plt.axis("off")
plt.imshow(wc)
```

```python
custom_mask = np.array(Image.open('./data/ny_nj.jpg'))
wc = WordCloud(background_color="black",
               stopwords=stopwords,
               mask = custom_mask,
               repeat=False,
               min_word_length=4,
              )
wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## Analyze an article using the 'bag of words' NLP technique

```python
# Initial imports
import nltk
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
article_df = open('./data/FemaWebDisasterDeclarations.csv', mode="r").read() 
```

```python
# Select and print the original single document text
doc_text = article_df
```

```python
# Creating the CountVectorizer instance defining the stopwords in English to be ignored
vectorizer = CountVectorizer(stop_words="english")

# The tokenization and occurrence counting
X = vectorizer.fit_transform([doc_text])

# Retrieve unique words list
words = vectorizer.get_feature_names()
print(words)
```

```python
# X raw data contains the occurrence of each term in the document. A unique ID is assigned to each term.
print(X)
```

```python
# The bag of words as DataFrame
words_df = pd.DataFrame(
    list(zip(words, np.ravel(X.sum(axis=1)))), columns=["Word", "Word_Count"]
)
words_df.head(21)
```

```python
# Creating the corpus 
corpus = [doc_text]

# Print sample document
print(corpus)
```

```python
# Obtaining the TF-IDF
vectorizer = TfidfVectorizer(stop_words="english")
X_corpus = vectorizer.fit_transform(corpus)
```

```python
# Obtaining matrix info
print(f"Matrix shape: {X_corpus.shape}")
print(f"Total number of documents: {X_corpus.shape[0]}")
print(f"Total number of unique words (tokens): {X_corpus.shape[1]}")
```

```python
# Retrieve words list from the corpus
words_corpus = vectorizer.get_feature_names()
print(words_corpus)
```

```python
# Obtaining the TF-IDF weight of each word in corpus as DataFrame
words_corpus_df = pd.DataFrame(
    list(zip(words_corpus, np.ravel(X_corpus.mean(axis=0)))), columns=["Word", "TF-IDF"]
)

words_corpus_df = words_corpus_df.sort_values(by=["TF-IDF"], ascending=False)
```

```python
# Highest 20 TF-IDF scores
words_corpus_df.head(21)
```

```python
# Lowest 20 TF-IDF scores
words_corpus_df.tail(21)
```
